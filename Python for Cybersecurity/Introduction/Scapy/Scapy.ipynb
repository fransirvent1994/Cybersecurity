For starters, what is Scapy?

Scapy is one of the Python libraries designed to work with network traffic. So, we´ll proceed with a demonstration below:

<h2>First step:</h2>
Install Python on your system.

<h2>Second step:</h2>
Import Scapy library from Python by using the terminal. Just write "python" on the terminal and program will be running.

<h3>Third step:<h2>
The program we´re gonna write on python must read a file that contains the network info. I´ve extracted this from wireshark webpage.
<a href="https://www.wireshark.org/resources#sample-captures">Link to samples from wireshark page</a> I extracted the sample from the http.cap especifically.

<h4>Running program:

```python
# First library is imported
from scapy.all import *

# Create a variable called "packet" that contains the order to write the file we want to read (in this case, is called http.cap)
packet = rdpcap("http.cap")

# To read some data this file contains, just write the variable to show it
packet

# After that, we can go deeper and create a new variable called p which value is the first value of the data
p = packet[0]

# We use the show function to show the data
p.show()
```

Below we have the progam complete and the results obtained. I´ve done it on my Kali terminal with Oracle VirtualBox.

It shows the first element data contained on the list of the file. We can continue exploring the rest of the data by using the same program.

```python
# Starting from the variable p which reads the first element on the list, we create a new one which reads the fourth element.

p4 = packet[3]

p4.show()
```

By using this we can read the data inside of a file extracted from a web-server or an app, getting vulnerable information.
